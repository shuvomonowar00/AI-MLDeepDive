# This notebook represents the complete PyTorch autograd implementation

```python
import torch
```

```python
x = torch.tensor(3.0, requires_grad=True)
```

```python
y = x ** 2
```

```python
z = y + 2 * x
```

```python
x
```

```python
y
```

```python
z
```

```python
z.backward()
```

```python
x.grad
```

```python
import torch.nn as nn
rnn = nn.LSTM(10, 20, 2)
input = torch.randn(5, 3, 10)
h0 = torch.randn(2, 3, 20)
c0 = torch.randn(2, 3, 20)
output, (hn, cn) = rnn(input, (h0, c0))
```
