Understanding the building blocks Dataset and DataLoader primitives in PyTorch

```python
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import datasets, transforms
import pandas as pd
```

```python
# To check cuda is available or not
torch.cuda.is_available()
```

Using a built-in dataset

```python
# Transform: convert image → tensor
transform = transforms.ToTensor()

# Download MNIST dataset
train_dataset = datasets.MNIST(
    root="data",
    train=True,
    transform=transform,
    download=True
)

# Create the dataloader
train_loader = DataLoader(
    dataset=train_dataset,
    batch_size=32,
    shuffle=True  
)

# Check one batch
images, labels = next(iter(train_loader))
print(images.shape)
print(labels.shape)
```

Custom dataset

```python
class MyDataset(Dataset):
    def __init__(self, csv_file):
        self.data = pd.read_csv(csv_file)
    
    def __len__(self):
        return len(self.data)
    
    def __getitem__(self, index):
        x = self.data.iloc[index, :-1].values
        y = self.data.iloc[index, -1]

        x = torch.tensor(x, dtype=torch.float32)
        y = torch.tensor(y, dtype=torch.float32)

        return x, y
```
