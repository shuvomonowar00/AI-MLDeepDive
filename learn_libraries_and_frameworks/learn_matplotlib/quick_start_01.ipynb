```python
import matplotlib.pyplot as plt
import numpy as np
```

A simple example

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4], [1,4,2,3])
plt.show()
```

Parts of a Figure

```python
fig = plt.figure()
fig, ax = plt.subplots()
fig, axs = plt.subplots(2, 2)
fig, axs = plt.subplot_mosaic([['left', 'right_top'],['left', 'right_bottom']])
```
