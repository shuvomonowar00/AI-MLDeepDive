# **üß† Step 1: Setup**

```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
```

```python
# Create input tensor (values from -5 to 5)
x = torch.linspace(-5, 5, 100)
x
```

```python
x.ndim
```

# **‚öôÔ∏è Step 2: Apply Activation Functions**

```python
# Different activations
sigmoid_y = torch.sigmoid(x)            # Sigmoid
tanh_y = torch.tanh(x)                  # Tanh
relu_y = F.relu(x)                      # ReLU
leaky_relu_y = F.leaky_relu(x, 0.1)     # Leaky ReLU
softmax_y = F.softmax(x, dim=0)         # Softmax (dim=0 for 1D input)
```

```python
sigmoid_y
```

# **üé® Step 3: Plot and Visualize**

```python
plt.figure(figsize=(10, 7))

plt.subplot(2, 3, 1)
plt.plot(x, x)
plt.title("Linear (No Activation)")
plt.grid(True)

plt.subplot(2, 3, 2)
plt.plot(x, sigmoid_y)
plt.title("Sigmoid")
plt.grid(True)

plt.subplot(2, 3, 3)
plt.plot(x, tanh_y)
plt.title("Tanh")
plt.grid(True)

plt.subplot(2, 3, 4)
plt.plot(x, relu_y)
plt.title("ReLU")
plt.grid(True)

plt.subplot(2, 3, 5)
plt.plot(x, leaky_relu_y)
plt.title("Leaky ReLU")
plt.grid(True)

plt.subplot(2, 3, 6)
plt.plot(x, softmax_y)
plt.title("Softmax (1D)")
plt.grid(True)

plt.tight_layout()
plt.show()
```

# **üß© Step 5: Using Activation Functions in a Model**

```python
import torch.nn as nn

# Example model
model = nn.Sequential(
    nn.Linear(4, 8),     # input ‚Üí hidden
    nn.ReLU(),           # activation
    nn.Linear(8, 3),     # hidden ‚Üí output
    nn.Softmax(dim=1)    # output activation for 3-class classification
)

# Sample input
x = torch.randn(2, 4)  # batch of 2 samples, 4 features each
output = model(x)

print("Model output:\n", output)
print("Sum of probabilities for each sample:", output.sum(dim=1))
```
