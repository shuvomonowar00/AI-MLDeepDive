Matrix transposing and reshaping

```python
arr3 = np.arange(6).reshape(2,3) # Reshape the array to (2,3) matrix
print(f"After Reverse: {arr3}")
# arr4 = arr3.transpose() # Transpose the matrix
arr4 = arr3.T 
print(f"After Transpose: {arr4}")
```

Matrix reshapping and flipping

```python
arr5 = np.arange(16).reshape(4,4) # Reshape the array to (4,4) matrix
print(arr5)
arr6 = np.flip(arr5)
print(f"After flipping all axis: {arr6}")
arr7 = np.flip(arr5, axis=1)
print(f"After flipping axis 1: {arr7}")
```
