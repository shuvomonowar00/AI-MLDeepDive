```python
from sklearn.datasets import make_classification
import torch
from torch.utils.data import Dataset, DataLoader
```

```python
X, y = make_classification(n_samples=10, n_features=2, n_informative=2, n_redundant=0, n_classes=2, random_state=42)
```

```python
X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.long)
```

```python
X
```

```python
y
```

```python
class CustomDataset(Dataset):
  def __init__(self, features, levels):
    self.features = features
    self.levels = levels

  def __len__(self):
    return self.features.shape[0]

  def __getitem__(self, index):
    return self.features[index], self.levels[index]
```

```python
dataset = CustomDataset(X, y)
```

```python
len(dataset)
```

```python
dataset[2]
```

```python
dataloader = DataLoader(dataset, batch_size=2, shuffle=False)
```

```python
for batch_features, batch_lavels in dataloader:
  print(batch_features)
  print(batch_lavels)
  print("-"*30)
```
