One Hot Encoder using Scikit-Learn

```python
import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

```python
data = np.array([['Red'], ['Blue'], ['Green']])
encoder = OneHotEncoder()
encoded = encoder.fit_transform(data).toarray().astype('int32')
print(encoded)
```

Get feature names (very important)

```python
features_name = encoder.get_feature_names_out()
```

Convert to pandas DataFrame

```python
import pandas as pd

encoded_df = pd.DataFrame(encoded, columns=features_name)
encoded_df
```

Avoid the Dummy Variable Trap (drop)

```python
encoder = OneHotEncoder(drop='first')
encoded = encoder.fit_transform(data)
print(encoded.toarray())
```

Handling unknown categories (REAL-WORLD IMPORTANT)

```python
encoder = OneHotEncoder(handle_unknown='ignore')
```
